```python
# Two basci data structure in pandas
# 1. Series
# Consist of 4 parts: data, index, dtype, name
s = pd.Series(data = [100, 'a', {'dic':5}],
              index = pd.Index(['id1', 20, 'third'], name = 'my_idx'),
              dtype = 'object',
              name = 'my_name')
s
```

```python
# 2: DataFrame
data = [
    [1,'a',1.2],
    [2, 'b', 2.2],
    [3, 'c', 3.2]
]
# [表达式 for 变量 in 可迭代对象]
df = pd.DataFrame(data = data,
                  index = ['row_%d' %i for i in range(3)],
                  columns=['col_0', 'col_1', 'col_2'] )
df
```

```python
# Second way to create a dataframe
# 
df2 = pd.DataFrame(
    data = {
        'col_0':[1,2,3],
        'col_1':list('abc'),
        'col_2':[1.2, 2.2, 3.2]
    },
    index = [f'row_{i}' for i in range(3)]
)
df2
```

```python
# Retrieve the coresponding columns and the table consisting of several columns
df2['col_0']
```

```python
df2[['col_0', 'col_1']]
```

```python
# df.values, retrieve the attributes
df2.values
```

```python
# Assign values to DataFrame without index and column
# By using dic, column names and index are assigned automatically
df3 = pd.DataFrame({
    'col_0':['a','b','c'],
    'col_1':[1, 2, 3],
    'col_2':["一","二", "三"]
})

df3
```

```python
# Normally, we assign column name and values

data = [['alice', 'bob', 'jack'], [2000, 5000, 3000], [23, 25, 27]]
df4 = pd.DataFrame(data,
                   columns = ['name', 'salary', 'age']
)
df4
```

```python
# !!! previous data block seems read data by row!
data = [['alice', 23, 2000], ['jack', 25, 3000], ['bob', 24, 5000]]
df5 = pd.DataFrame(
    data,
    columns = ['name','age','salary']
)
df5
```
