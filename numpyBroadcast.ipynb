```python
# The broadcast mechanism is used to handle operations between two arrays of
# different dimensions
import numpy as np
import pandas as pd
```

```python
# scalar and an array perform operations
res = 3 * np.ones((2,2)) + 1
res
```

```python
res = 1 / (res)
```

```python
res
```

```python
# computation between vector with matrix
# dot
# array.dot(array)
a = np.array([1,2,3])
b = np.array([1,3,5])
a.dot(b)
```

```python
# vector norm
exp_array = np.array([3,4])
np.linalg.norm(exp_array)
```

```python
# in matrix norm, ord parameter is the most important
# ord:
# - None Frobenius norm
# - fro Frobenius norm
# - nuc nuclear norm
# - inf max(sum(abs(x)), axis=1)
matrix_target = np.arange(4).reshape(-1,2)
matrix_target
```

```python
np.linalg.norm(matrix_target, 'fro')
```

```python
np.linalg.norm(matrix_target, 1)
```
