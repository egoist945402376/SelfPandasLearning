```python
# transpose
# np.array.T
np.zeros((2,3)).T
```

```python
np.empty((3,4)).T
```

```python
# combination operation
# r_, c_
# usage: np.r_[np.array_1, np.array_2]
np.r_[np.zeros((2,3)), np.zeros((2,3))]
```

```python
np.c_[np.zeros((2,3)),np.zeros((2,3))]
```

```python
# numpy.reshape
# Rearrange the array according to a new dimension, with two mode C and F
# C: Filled by row
# F: Filled by column
np.arange(8).reshape((2,4))
```

```python
np.arange(8).reshape((2,4), order='C')
```

```python
np.arange(8).reshape((2,4),order='F')
```

```python
# Since the dimensions of the array is confirmed, there can be a dimension empty, filled by -1
np.arange(9).reshape((3,-1), order='F')
```

```python
# Transform n*1 array to one dimension array
np.zeros((3,1)).reshape(-1)
```

```python
# slicing in numpy
target = np.arange(9).reshape(3,3)
target[:-1,[0,2]]
```

```python
# When two are both list (advanced index), pairwise matching
target[[0,1], [0,2]]
```

```python
# utilize np.ix_ to boolean index on corresponding dimension
target[np.ix_([True, False, True], [True, False, True])]
```

```python
# For arrays that contain missing values, to skip the missing values, you need to use
# nan* functions.
target = np.array([1, 2, np.nan, np.nan])
```

```python
target.max()
np.nanmax(target)
```

```python
np.nanmax(target)
```

```python
np.nanquantile(target,0.5)
```

```python
# for covirance and correlation
target1= np.array([1, 3, 5, 9])
target2 = np.array([1,5,3,-9])
```

```python
# np.cov(array1, array2)
np.cov(target1, target2)
```

```python
np.corrcoef(target1, target2)
```

```python
# For 2d array, statistic functions' aixs parameter means calculate the stat on a given dimension
# axis = 0 -> column
# aixs = 1 -> row
target = np.arange(1,10).reshape(3,-1)
target
```

```python
target.sum(axis=0)
```

```python
target.sum(axis=1)
```
