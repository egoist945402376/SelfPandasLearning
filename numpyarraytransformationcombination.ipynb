```python
import numpy as np
import pandas as pd
```

```python
# transpose
# np.array.T
np.zeros((2,3)).T
```

```python
np.empty((3,4)).T
```

```python
# combination operation
# r_, c_
# usage: np.r_[np.array_1, np.array_2]
np.r_[np.zeros((2,3)), np.zeros((2,3))]
```

```python
np.c_[np.zeros((2,3)),np.zeros((2,3))]
```

```python
np.array([0,0])
```

```python
np.zeros(2)
```

```python
# numpy.reshape
# Rearrange the array according to a new dimension, with two mode C and F
# C: Filled by row
# F: Filled by column
np.arange(8).reshape((2,4))
```

```python
np.arange(8).reshape((2,4), order='C')
```

```python
np.arange(8).reshape((2,4),order='F')
```

```python
# Since the dimensions of the array is confirmed, there can be a dimension empty, filled by -1
np.arange(9).reshape((3,-1), order='F')
```

```python
# Transform n*1 array to one dimension array
np.zeros((3,1)).reshape(-1)
```

```python
# slicing in numpy
target = np.arange(9).reshape(3,3)
target[:-1,[0,2]]
```

```python
# When two are both list (advanced index), pairwise matching
target[[0,1], [0,2]]
```

```python
# utilize np.ix_ to boolean index on corresponding dimension
target[np.ix_([True, False, True], [True, False, True])]
```
