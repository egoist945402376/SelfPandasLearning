```python
L = []
```

```python
def my_func(x):
    return x*2
```

```python
for i in range(5):
    L.append(my_func(i))
```

```python
L
```

```python
# [* for i in *], first i: map function, second i: iterable
[my_func(i) for i in range(5)]
```

```python
[m+'_'+n for m in ['a','b'] for n in ['c','d']]
```

```python

# condition else b
value = 'cat' if 2>1 else 'dog'
value
```

```python
# Intercept list
L = [1,2,3,4,5,6,7,8,9]
```

```python
[i if i <= 5 else 5 for i in L]
```

```python
# Anonymous function (lambda)
my_func = lambda x: x*2
a = my_func(6)
a
```

```python
multi_para_func = lambda a,b:a+b
b = multi_para_func(2,3)
b
```

```python
[(lambda x: x*2)(i) for i in range(5)]
```

```python
# map
list(map(lambda x: x**2, range(5)))
```

```python
list(map(lambda x,y:x+y,range(5), range(5)))
```
